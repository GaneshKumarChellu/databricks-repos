```sql
create schema if not exists demo_db
```

```sql
create table if not exists databricks_ws.demo_db.fire_service_calls_tbl(
  CallNumber integer,
  UnitID string,
  IncidentNumber integer,
  CallType string,
  CallDate string,
  WatchDate string,
  CallFinalDisposition string,
  AvailableDtTm string,
  Address string,
  City string,
  Zipcode integer,
  Battalion string,
  StationArea string,
  Box string,
  OriginalPriority string,
  Priority string,
  FinalPriority integer,
  ALSUnit boolean,
  CallTypeGroup string,
  NumAlarms integer,
  UnitType string,
  UnitSequenceInCallDispatch integer,
  FirePreventionDistrict string,
  SupervisorDistrict string,
  Neighborhood string,
  Location string,
  RowID string,
  Delay float
) using DELTA
```

```sql
insert into demo_db.fire_service_calls_tbl
select * from global_temp.fire_service_calls_view
```

```sql
Select count(*) from demo_db.fire_service_calls_tbl;
```

```sql
Select CallDate, date_format(CallDate,"dd/MM/yyyy") from demo_db.fire_service_calls_tbl limit 100
```

```sql
Select distinct YEAR(CallDate) from demo_db.fire_service_calls_tbl
```

```sql
Select weekofyear(to_date(CallDate)) as week_no, count(*) as Call_count from demo_db.fire_service_calls_tbl where year(to_date(CallDate)) = 2018
group by 1 order by 2 desc
```

```sql
Select * from demo_db.fire_service_calls_tbl limit 10
```

```sql
Select Neighborhood, SUM(Delay) from demo_db.fire_service_calls_tbl where year(to_date(CallDate)) = 2018 group by 1 order by 2 desc
```
